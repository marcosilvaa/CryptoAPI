```python
btc = md.price_data(symbol='BTCUSDT', interval='D', start='2015-01-01', category='spot')
btc['returns'] = btc.Close.pct_change()
#btc['returns'] = (btc.Close.pct_change()*100).round(4)
btc.dropna(inplace=True)
```

```python
btc['log_returns'] = np.log(btc['returns'])
```

```python
btc
```

```python
btc.returns.hist(bins=100)
```

```python
btc.log_returns.hist(bins=100)
```

```python
btc['log_sum'] = btc['log_returns'].cumprod()
```

```python
btc['log_sum'].plot()
```

```python
btc.Close.plot()
```

```python
log_returns  = np.log(btc.Close/btc.Close.shift(1)).dropna()
log_returns.cumsum().plot()
```
